```python
import requests
import pandas as pd
import sqlite3
import lxml.html
```

```python
r = requests.get("https://en.wikipedia.org/wiki/List_of_Nintendo_Entertainment_System_games")
```

```python
html                = lxml.html.fromstring(r.text)
software_list_table = html.get_element_by_id("softwarelist")
tablebody           = software_list_table.getchildren()[0]
```

```python
#setting up a pandas dictionary
data = {
    "title":            [],
    "release_date_NA":  [],
    "release_date_PAL": [], 
    "publisher":        [], 
    "developer":        []

}

for row in tablebody.getchildren()[2:]:
    cells = row.getchildren()
    
    data["title"].append(            cells[0].text_content().strip() )
    data["release_date_NA"].append(  cells[1].text_content().strip() )
    data["release_date_PAL"].append( cells[2].text_content().strip() )
    data["publisher"].append(        cells[3].text_content().strip() )
    
    if len(cells) < 5:    
        data["developer"].append("Unlisted")
    else:
        data["developer"].append(    cells[4].text_content().strip() )
    
df = pd.DataFrame(data)
df
```

```python
db     = sqlite3.connect(":memory:")
cursor = db.cursor()
cursor.execute("""
    CREATE TABLE GAMES(TITLE, RELEASE_DATE_NA, RELEASE_DATE_PAL, PUBLISHER, DEVELOPER)
""")

for row in df[df.publisher.isin(["Konami", "Capcom"])].itertuples():
    insert_sql_syntax =  """
        INSERT INTO GAMES(TITLE, RELEASE_DATE_NA, RELEASE_DATE_PAL, PUBLISHER, DEVELOPER)
        VALUES (?,?,?,?,?)
    """
    cursor.execute(insert_sql_syntax, row[1:])
db.commit()
# db.close()
```
